# Validación cruzada k-fold

---

### **¿Qué es la validación cruzada?**
La validación cruzada es una técnica para evaluar cómo de bien funcionará un modelo de machine learning en datos nuevos (que no ha visto antes). Su objetivo es comprobar si el modelo generaliza bien o si solo está aprendiendo de memoria los datos de entrenamiento.

---

### **¿Qué es k-fold cross-validation?**

1. **Divide tus datos en \(k\) partes:**
   - Imagina que tienes un conjunto de datos.
   - Lo divides en \(k\) partes de tamaño similar (llamadas "folds"). Por ejemplo, si \(k = 5\), divides los datos en 5 partes iguales.

2. **Entrena y prueba varias veces:**
   - En cada iteración, usa \(k-1\) partes para entrenar el modelo y la parte restante para probarlo. Esto se repite \(k\) veces, cambiando cada vez qué parte usas para probar.
   - Por ejemplo:
     - Iteración 1: Entrena en los folds 2, 3, 4, 5; prueba en el fold 1.
     - Iteración 2: Entrena en los folds 1, 3, 4, 5; prueba en el fold 2.
     - Y así sucesivamente.

3. **Promedia los resultados:**
   - Al final, tienes \(k\) mediciones de rendimiento (una por cada fold).
   - Calculas el promedio de esas mediciones para tener una estimación más confiable de cómo rendirá tu modelo en datos nuevos.

---

### **¿Por qué es útil?**

1. **Evita sobreajuste y subajuste:**
   - Permite evaluar el modelo en diferentes partes de los datos.
   - Ayuda a detectar si el modelo está aprendiendo demasiado de memoria o si no está capturando bien los patrones.

2. **Mejor uso de los datos:**
   - En lugar de apartar un gran conjunto de datos solo para validación, cada dato se usa tanto para entrenar como para probar (en diferentes iteraciones).

3. **Resultados más confiables:**
   - Promediar el rendimiento en \(k\) diferentes conjuntos de validación da una evaluación más robusta.

---

### **Ejemplo práctico (5-fold cross-validation):**
Imagina que tienes 100 datos y eliges \(k = 5\). Divides tus datos en 5 grupos:

1. Iteración 1:
   - Entrenamiento: 80 datos (folds 2, 3, 4, 5).
   - Prueba: 20 datos (fold 1).

2. Iteración 2:
   - Entrenamiento: 80 datos (folds 1, 3, 4, 5).
   - Prueba: 20 datos (fold 2).

3. Iteración 3:
   - Entrenamiento: 80 datos (folds 1, 2, 4, 5).
   - Prueba: 20 datos (fold 3).

4. Iteración 4:
   - Entrenamiento: 80 datos (folds 1, 2, 3, 5).
   - Prueba: 20 datos (fold 4).

5. Iteración 5:
   - Entrenamiento: 80 datos (folds 1, 2, 3, 4).
   - Prueba: 20 datos (fold 5).

Al final, calculas el promedio del rendimiento de las 5 pruebas.

---

### **¿Qué valor de \(k\) elegir?**
- **\(k = 5\) o \(k = 10\):** Son valores comunes. Ofrecen un buen balance entre tiempo de cómputo y calidad de la evaluación.
- **\(k = número de datos\):** Se llama **Leave-One-Out Cross-Validation (LOOCV)**. Es muy preciso, pero computacionalmente costoso.

---

En resumen, **k-fold cross-validation** es una forma inteligente de usar tus datos al máximo para entrenar y validar tu modelo de forma más confiable. Es como probar a tu modelo en diferentes escenarios antes de confiar en él. ¿Te gustaría ver un ejemplo implementado?
