# Seleccin de caractersticas

## Mtodos de filtrado: 
El mtodo de filtrado para la seleccin de caractersticas es una tcnica utilizada para reducir el nmero de caractersticas (o variables) en un conjunto de datos, dejando solo las que son ms relevantes para el problema que estamos resolviendo.

##### Qu es la seleccin de caractersticas?
En machine learning, a veces tienes demasiadas caractersticas (columnas en tus datos), y no todas son tiles. Algunas pueden ser irrelevantes o redundantes. El objetivo de la seleccin de caractersticas es encontrar el conjunto ms pequeo de caractersticas importantes que an permitan al modelo desempearse bien.

##### Qu hace el mtodo de filtrado?
El mtodo de filtrado utiliza criterios estadsticos para evaluar cada caracterstica de forma independiente del modelo que usars despus. Esto significa que se enfoca nicamente en las caractersticas y su relacin con la variable objetivo, sin construir un modelo an.

En palabras simples, clasifica las caractersticas segn su importancia y selecciona las mejores basndose en una mtrica.



### Existen:
* Mtodos de filtrado para regresin
* Mtodos de filtrado para clasificacin
* Mtodos de filtrado en un problema realista
* Mtodos wrapper
* Mtodos embedded

### Ventajas e inconvenientes


##### Ventajas del mtodo de filtrado:
* Rpido y sencillo: no necesitas entrenar un modelo para aplicar este mtodo.
* Independiente del modelo: puedes usarlo antes de probar diferentes algoritmos, lo que lo hace muy verstil.
* Evita overfitting: reducir caractersticas irrelevantes o redundantes ayuda a que el modelo generalice mejor.

##### Desventajas:
* Ignora interacciones entre caractersticas: como evala cada caracterstica por separado, puede pasar por alto combinaciones importantes de caractersticas.
* Puede ser demasiado simple: si las relaciones entre caractersticas y la variable objetivo son complejas, este mtodo puede no capturarlas.

## Conceptos Importantes:

### 1. F-Test
El F-Test (prueba F) es una tcnica estadstica que se utiliza para comparar la relacin entre dos variables: una caracterstica y la variable objetivo (lo que queremos predecir). Sirve para evaluar si una caracterstica es estadsticamente significativa al predecir la variable objetivo.

#### Cmo funciona?
1. Mide la relacin entre la variabilidad dentro de los grupos (valores de la caracterstica) y la variabilidad entre los grupos.
2. El resultado es un puntaje F. Cuanto mayor sea, ms fuerte es la relacin entre la caracterstica y la variable objetivo.

#### Para qu sirve en machine learning?
* Ayuda a seleccionar caractersticas relevantes en problemas de regresin o clasificacin.
* til cuando trabajas con datos continuos (p. ej., edad, peso, ingresos).
* Por ejemplo:
    * Si ests prediciendo los precios de casas, el F-Test puede indicar que "tamao de la casa" tiene una relacin ms fuerte con el precio que "color de las paredes".

### 2. Mutual Information Score (MI)
El Mutual Information Score (MI) mide la dependencia entre dos variables. En otras palabras, indica cunto una caracterstica reduce la incertidumbre sobre la variable objetivo.

#### Cmo funciona?
1. Calcula cunta informacin aporta una caracterstica sobre la variable objetivo.
2. El resultado es un puntaje MI. Un valor alto significa que la caracterstica es muy relevante para predecir la variable objetivo.

#### Para qu sirve en machine learning?
* Se usa tanto en problemas de clasificacin como de regresin.
* Funciona bien con variables continuas y categricas.
* Por ejemplo:
    * Si ests prediciendo si un correo es spam o no, el MI podra revelar que la presencia de ciertas palabras ("gratis", "oferta") tiene una fuerte relacin con la clasificacin de "spam".



### Conclusin

* Usa el F-Test cuando tus datos y relaciones sean principalmente lineales y trabajes con variables continuas.
* Usa Mutual Information Score (MI) para detectar relaciones ms complejas y no lineales, especialmente si tienes variables categricas.

Ambos son herramientas valiosas para seleccionar caractersticas y mejorar el rendimiento de tus modelos de machine learning.
