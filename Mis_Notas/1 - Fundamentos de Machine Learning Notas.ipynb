# K-Nearest Neighbors (KNN)

El algoritmo **K-Nearest Neighbors (KNN)** es uno de los algoritmos más simples y efectivos en machine learning para tareas de clasificación y regresión. Aquí tienes una explicación sencilla:

### **¿Cómo funciona KNN?**

KNN clasifica o predice un valor basándose en los datos más cercanos a un punto de interés. En esencia, **"mira a sus vecinos"** para decidir qué etiqueta asignar o qué valor predecir.

1. **Distancia a los vecinos:** 
   - Para un nuevo dato, KNN calcula la distancia entre este punto y todos los puntos en el conjunto de datos.
   - Las distancias más comunes son:
     - Distancia Euclidiana.
     - Distancia Manhattan.
     - Distancia Coseno.

2. **Seleccionar los \(k\) vecinos más cercanos:**
   - Ordena los datos por distancia y selecciona los \(k\) puntos más cercanos.
   - El valor de \(k\) se define antes de ejecutar el algoritmo (p. ej., \(k=3\)).

3. **Decidir la predicción:**
   - **Para clasificación:** Se asigna la etiqueta que aparece con más frecuencia entre los \(k\) vecinos (votación de mayoría).
   - **Para regresión:** Se calcula la media de los valores de los \(k\) vecinos.

---

### **Ventajas de KNN:**

1. **Simple y fácil de entender:** No requiere un modelo complejo ni entrenamiento; solo almacena los datos.
2. **No paramétrico:** No asume ninguna distribución en los datos.
3. **Versátil:** Funciona bien para clasificación y regresión.

---

### **Desventajas de KNN:**

1. **Lento en predicción:** Si tienes muchos datos, calcular distancias para cada punto puede ser costoso.
2. **Sensible a los valores de \(k\):**
   - Un \(k\) demasiado bajo puede llevar a sobreajuste.
   - Un \(k\) demasiado alto puede causar subajuste.
3. **Dependencia en la escala de las características:**
   - Si las variables tienen escalas diferentes, las distancias pueden distorsionar los resultados. Es importante normalizar o estandarizar los datos.
4. **No maneja bien datos muy dispersos o de alta dimensionalidad.**

---

### **Ejemplo práctico:**
Imagina que tienes un conjunto de datos con frutas, donde cada fruta tiene características como peso y color. Quieres predecir si una nueva fruta es una manzana o una naranja.

1. Usas KNN para calcular las distancias de esta nueva fruta a todas las frutas en tu conjunto de datos.
2. Seleccionas los \(k\) vecinos más cercanos (p. ej., las \(3\) frutas más cercanas).
3. Si la mayoría de esos vecinos son manzanas, clasificas la fruta como una manzana.


### **¿Cómo eleger K?:**

Elegir un buen valor para \(k\) es crucial para el desempeño del algoritmo **K-Nearest Neighbors (KNN)**. Si \(k\) es demasiado bajo o demasiado alto, el modelo puede sufrir de **sobreajuste** o **subajuste**, respectivamente. Aquí tienes las principales estrategias para elegir \(k\):

---

### **1. Consideraciones generales:**
- **\(k = 1\):** El modelo se ajusta demasiado a los datos (sobreajuste). Cada punto nuevo se clasifica exactamente como el punto más cercano, lo que puede llevar a errores debido al ruido.
- **\(k\) muy alto:** El modelo puede subajustarse porque considera demasiados puntos, lo que diluye la influencia de los vecinos más cercanos.

Generalmente, \(k\) se elige como un valor **impar** para evitar empates (en problemas de clasificación binaria).

---

### **2. Estrategias para elegir \(k\):**

#### **a. Probar diferentes valores y validar:**
- Utiliza un conjunto de validación (o validación cruzada) para probar varios valores de \(k\).
- Mide el rendimiento del modelo para cada \(k\) (por ejemplo, precisión, F1-score, o error promedio).
- Selecciona el \(k\) que minimice el error en los datos de validación.

#### **b. Regla empírica:**
- Una regla común es elegir \(k\) como la raíz cuadrada del tamaño del conjunto de datos (\(n\)):
  \[
  k = \sqrt{n}
  \]
  Redondea al entero más cercano. Esto funciona como un buen punto de partida.

#### **c. Considerar el tamaño del conjunto de datos:**
- Si tienes un conjunto de datos pequeño, usa valores de \(k\) más bajos (por ejemplo, entre 1 y 10).
- Si tienes un conjunto de datos grande, puedes probar valores más altos, pero no tan altos como para incluir demasiados puntos y diluir la decisión.

---

### **3. Evaluación del impacto de \(k\):**

#### **Cuando \(k\) es bajo (por ejemplo, \(k = 1\)):**
- Ventajas:
  - Muy sensible a los patrones locales.
  - Útil para datos donde los puntos están bien separados.
- Desventajas:
  - Muy sensible al ruido.
  - Puede clasificar incorrectamente debido a outliers.

#### **Cuando \(k\) es alto (por ejemplo, \(k > 20\)):**
- Ventajas:
  - Reduce la influencia del ruido.
  - Es menos probable que se vea afectado por outliers.
- Desventajas:
  - Ignora patrones locales.
  - Puede causar subajuste, ya que el modelo se convierte en una aproximación muy general.

---

### **4. Validación cruzada (Cross-Validation):**
Usa **validación cruzada** para encontrar el mejor valor de \(k\):
1. Divide tus datos en \(k\) subconjuntos (folds).
2. Ajusta y evalúa el modelo para diferentes valores de \(k\).
3. Elige el valor de \(k\) que tenga el mejor rendimiento promedio en los datos de validación.

---

### **Ejemplo Práctico:**
Supongamos que tienes 100 datos de entrenamiento. Podrías:
1. Probar valores de \(k = 1, 3, 5, 7, 10, \sqrt{100} = 10\), etc.
2. Medir el rendimiento para cada \(k\) utilizando un conjunto de validación.
3. Elegir el \(k\) con el menor error de validación.

---
