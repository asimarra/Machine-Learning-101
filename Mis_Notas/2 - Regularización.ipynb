## Regularización

### Parámetro de regularización

- Ridge es un buen punto de partida para analizar un problema: por defecto, usarlo.
- Lasso es recomendado para la extracción de carácteristicas. Hace el módelo mas simple. (Selecciona carácteristicas que son realmente útil).
- Recuerda, siempre estamos buscando valores alpha que esten mas cercas de cero.
- La unica diferencia con Ridge a nivel de formula es que ||W|| no esta elevado al cuadrado.
- Aumentar Alpha aumenta la regularización.
